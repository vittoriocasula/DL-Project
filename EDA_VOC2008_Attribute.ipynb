```python
df_train = pd.read_csv(root + "data/VOC2008_attribute/ann/ann_train.txt", delimiter=" ")

attr_name = pd.read_csv(
    root + "data/VOC2008_attribute/attribute_names.txt", delimiter="\t", header=None
)
# print(*attr_name[0].unique(), sep="\n")

attr_name = attr_name.values.tolist()

tmp = []
for list in attr_name:
    for str in list:
        tmp.append(str)

attr_name = ["img", "class", "xmin", "ymin", "xmax", "ymax"] + tmp
attr_name

df_train.columns = attr_name
del df_train["img"]
del df_train["class"]
del df_train["xmin"]
del df_train["ymin"]
del df_train["xmax"]
del df_train["ymax"]

# df_train.apply(pd.Series.value_counts)
# order by number of 1s (presence of attribute)
row_index = 1
df_train = df_train.apply(pd.Series.value_counts)
df_sorted_columns = df_train[df_train.columns[df_train.iloc[row_index].argsort()]]

frequencies = df_sorted_columns.loc[1]

plt.figure(figsize=(12, 6))
plt.bar(frequencies.index, frequencies.values)

plt.xlabel("Attributes", fontsize=10)
plt.ylabel("Frequency", fontsize=10)
plt.title("Attributes Distribution - Train Split", fontsize=14)

plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=10)
plt.grid(axis="y", linestyle="--", linewidth=0.7)

plt.show()
```
